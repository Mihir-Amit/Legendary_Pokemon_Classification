```python
data = pd.read_csv('/content/drive/MyDrive/ML DATASETS/Pokemon Classification/Pokemon.csv')
data.head()
```

```python
data.drop(columns=['#','Name',	'Type 1',	'Type 2'],inplace=True)
```

```python
data
```

```python
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
data['Legendary']= label_encoder.fit_transform(data['Legendary'])

data['Legendary'].unique()
```

```python
data.Legendary.value_counts()
```

```python
from sklearn.model_selection import train_test_split

pokemon_features = data.drop("Legendary",axis=1)
target = data["Legendary"]

X_train,X_test,Y_train,Y_test = train_test_split(pokemon_features,target,test_size=0.20,random_state=0)
```

```python
from sklearn.metrics import accuracy_score
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
model = Sequential()
model.add(Dense(32,activation='relu',input_dim=8))
model.add(Dense(8, activation='sigmoid'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=100, callbacks = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3))
```

```python
Y_pred_nn = model.predict(X_test)
rounded = [round(x[0]) for x in Y_pred_nn]
Y_pred_nn = rounded
score_nn = round(accuracy_score(Y_pred_nn,Y_test)*100,2)
print("The accuracy score achieved is: "+str(score_nn)+" %")
```
